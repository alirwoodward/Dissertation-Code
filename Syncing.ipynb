# Looking for beginning spikes- Tuesday

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
movement_data = pd.read_csv('movement_data.csv')

# Filter the data for the first 15 minutes (900 seconds)
first_15_minutes = movement_data[movement_data['Time'] <= 900]

# Set the figure size
plt.figure(figsize=(15, 8))

# Plot each participant's movement data
movement_columns = movement_data.columns[3:]
offset = 0
for column in movement_columns:
    plt.plot(first_15_minutes['Time'], first_15_minutes[column] + offset, label=column)
    offset += 1  # Increase the offset for the next participant

plt.title('Movement Data of Participants Over the First 15 Minutes (Spread Out)')
plt.xlabel('Time (seconds)')
plt.ylabel('Movement Value (with offset)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

# Creating new CSV files that only contain the data that correlates to the times in the video- Tuesday

```python
# Load the datasets and filter out rows where "Time" is less than 5664
import pandas as pd

# Load the datasets
axw_data = pd.read_csv('axw_data.csv')
movement_data = pd.read_csv('movement_data.csv')
awc_data = pd.read_csv('awc_data.csv')

# delete all data that is before the video starts
axw_video_data = axw_data[axw_data['Time'] >= 5664]
del axw_video_data[axw_video_data.columns[0]]

movement_video_data = movement_data[movement_data['Time'] >= 5664]
del movement_video_data[movement_video_data.columns[0]]

awc_video_data = awc_data[awc_data['Time'] >= 5664]
del awc_video_data[awc_video_data.columns[0]]
```

# Finding a moment in the data for syncing to Friday video

```python
# Filter the data for the time range 200-300 seconds
zoomed_data = movement_data[(movement_data['Time'] >= 5021) & (movement_data['Time'] <= 5141)]

# Set the figure size
plt.figure(figsize=(15, 8))

df = movement_data[['ER', 'P11']]

movement_columns = df.columns[0:]
offset = 0
for column in movement_columns:
    plt.plot(zoomed_data['Time'], zoomed_data[column] + offset, label=column)
    offset += 10  # Increase the offset for the next participant

# Formatting the plot
plt.title('Movement Data of Participants P11ER (Seconds 5021-5141)')
plt.xlabel('Time (seconds)')
plt.ylabel('Movement Value (with offset)')
plt.xticks(rotation=50)
plt.grid(True, alpha=0.3)
plt.xticks(range(5021, 5141, 2))

plt.legend()
plt.tight_layout()

plt.show()
```

The above plot shows the movement data that begins at minute 11:02 in the video. Particularly, the three orange spikes correlate to P11 performing swing movements with his arms that are the cause for the sudden spike. The first swing happens at 11:10 in the video (5029 in the data), swing two occurs at 11:32 (5051), swing three occurs at 11:51 (5070). Using this information the data will be synced with the seconds in the video.
