# Converting MATLAB data to new dataframes for python

```python
import scipy.io
import pandas as pd

# Load the MATLAB file
mat_file = 'Fri_wavelet_timeseries_results.mat'
#mat_file = 'Tues_wavelet_timeseries_results.mat'

mat_contents = scipy.io.loadmat(mat_file)

# Display the keys in the MATLAB file to understand its structure
print(mat_contents.keys())
```

## Create csv for AXW data

```python
import numpy as np
import pandas as pd


R = mat_contents['R'][0][0]
times = mat_contents['times'].flatten()

# Extract the 'axw' data
axw_data = R['axw']

# Create a DataFrame for the 'axw' data with all columns
axw_df = pd.DataFrame(axw_data.T)

# Add the 'times' column
axw_df['Time'] = times


# Move 'Time' column to the front
cols = axw_df.columns.tolist()
cols.insert(0, cols.pop(cols.index('Time')))
axw_df = axw_df[cols]

# Rename the columns
axw_df.columns = ['Time'] + [str(i) for i in range(1, len(axw_df.columns))]


# Display the head of the DataFrame
#print(axw_df.head())
#print(axw_df.describe())

axw_df.head()
```

```python
#axw_df['Time']
```

```python
axw_df.to_csv('fri_axw_data.csv')
```

## Create csv for AWC data

```python
# Extract the 'awc' data
awc_data = R['awc']

# Create a DataFrame for the 'awc' data with all columns
awc_df = pd.DataFrame(awc_data.T)

# Add the 'times' column
awc_df['Time'] = times

# Move 'Time' column to the front
cols = awc_df.columns.tolist()
cols.insert(0, cols.pop(cols.index('Time')))
awc_df = awc_df[cols]

# Rename the columns
awc_df.columns = ['Time'] + [str(i) for i in range(1, len(awc_df.columns))]

awc_df.head()

# Display the head of the DataFrame
#print(synchrony_df.head())
#print(synchrony_df.describe())
```

```python
awc_df.to_csv('fri_awc_data.csv')
```

## Create raw data csv that uses the same measure of time

```python
movement_df = pd.read_csv('Fri_processed_for_matlab_acc.csv')
#movement_df = pd.read_csv('Tues_processed_for_matlab_acc.csv')

# Convert the 'datetime' column to datetime type
movement_df['datetime'] = pd.to_datetime(movement_df['datetime'])

# Calculate the time in seconds from the start
start_time = movement_df['datetime'].iloc[0]
movement_df['Time'] = (movement_df['datetime'] - start_time).dt.total_seconds()

# Move 'Time' column to the front
cols = movement_df.columns.tolist()
cols.insert(0, cols.pop(cols.index('Time')))
movement_df = movement_df[cols]
```

```python
movement_df.head()
```

```python
movement_df.to_csv('fri_movement_data.csv')
```

# Creating CSV files for all of this data per second (less data for visualizations)

## AXW data per second

```python
# Create a new dataframe to store one data point for every second
axw_unique_seconds = np.arange(axw_df['Time'].min(), axw_df['Time'].max() + 1)
axw_closest_indices = [np.abs(axw_df['Time'] - sec).idxmin() for sec in axw_unique_seconds]
axw_per_second = axw_df.loc[axw_closest_indices].reset_index(drop=True)

axw_per_second.head()
```

```python
axw_per_second.to_csv('fri_axw_data_perSecond.csv')
```

```python
# Create a new dataframe to store one data point for every second
awc_unique_seconds = np.arange(awc_df['Time'].min(), awc_df['Time'].max() + 1)
awc_closest_indices = [np.abs(awc_df['Time'] - sec).idxmin() for sec in awc_unique_seconds]
awc_per_second = awc_df.loc[awc_closest_indices].reset_index(drop=True)

awc_per_second.head()
```

```python
awc_per_second.to_csv('fri_awc_data_perSecond.csv')
```
